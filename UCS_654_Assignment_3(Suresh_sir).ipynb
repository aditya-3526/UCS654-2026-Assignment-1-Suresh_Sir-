```python
import pandas as pd
import numpy as np
import math

df = pd.read_csv("data.csv", encoding="latin1", low_memory=False)
x = df["no2"].dropna().values.astype(float)

r = 102303526
a_r = 0.05 * (r % 7)
b_r = 0.3 * (r % 5 + 1)

print("a_r =", a_r)
print("b_r =", b_r)

x = x[(b_r * x >= -1) & (b_r * x <= 1)]

z = []
for xi in x:
    zi = xi + a_r * math.asin(b_r * xi)
    z.append(zi)

z = np.array(z)

mean_value = sum(z) / len(z)
variance = sum((zi - mean_value)**2 for zi in z) / len(z)

lambda_value = 1 / (2 * variance)
c_val = math.sqrt(lambda_value / math.pi)

print("mean =", mean_value)
print("variance =", variance)
print("lambda =", lambda_value)
print("c =", c_val)
```
